```python
import mglearn
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.decomposition import PCA
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('Raisin_Dataset.xlsx')
df
```

```python
x = df.drop(['Area','Class'],axis='columns')

y = df[['Class']]
```

```python
from sklearn.preprocessing import StandardScaler

StandardScaler = StandardScaler()
```

```python
df_mm = StandardScaler.fit_transform(x)
x_mm = pd.DataFrame(data= df_mm, columns=x.columns)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_mm,y, train_size=0.7, random_state=111)
```

```python
cls = KNeighborsClassifier(n_neighbors= 3, weights= 'uniform', metric='minkowski')
```

```python
cls.fit(X_train,y_train)
```

# 적절한 K개의 수를 만들기
- 엘보우차트

```python
k_cnt = range(1,31)

acc_score = []
for k in k_cnt:
    cls_ml1 = KNeighborsClassifier(n_neighbors= k, weights= 'uniform')
    cls_ml1.fit(X_train, y_train)
    acc_score.append(1 - cls_ml1.score(X_test, y_test)) #오분류율에 대한 계산

plt.plot(k_cnt, acc_score, 'o--', color='#7279a1')
plt.xlabel('k')
plt.ylabel('Misclassification Rate')
plt.ylim(0.10, 0.25)
plt.show()
```

```python
cls = KNeighborsClassifier(n_neighbors= 15, weights= 'uniform', metric='minkowski')
cls.fit(X_train, y_train)
```

```python
print(cls.score(X_train, y_train), 'train의 정확도')
print(cls.score(X_test, y_test),'test의 정확도')
```

```python
k_cnt = range(1,31)

acc_score = []
for k in k_cnt:
    cls_ml2 = KNeighborsClassifier(n_neighbors= k, weights= 'distance')
    cls_ml2.fit(X_train, y_train)
    acc_score.append(1 - cls_ml2.score(X_test, y_test))
plt.plot(k_cnt, acc_score, 'o--', color='#66d48e')

plt.xlabel('k')
plt.ylabel('Misclassification Rate')
plt.ylim(0.10, 0.25)

plt.show()
```

- 가중치 모델을 학습했을 때. k=30

```python
cls = KNeighborsClassifier(n_neighbors = 30, weights = 'distance')
cls.fit(X_train,y_train)

print(cls.score(X_train, y_train), 'train의 정확도')
print(cls.score(X_test, y_test),'test의 정확도')
# 과적합의 결과가 나왔다. k=30이라는 값이 적절하진 않음.
```

# K range를 늘려보자

```python
k_cnt = range(1,201)

acc_score = []
for k in k_cnt:
    cls_ml2 = KNeighborsClassifier(n_neighbors= k, weights= 'distance')
    cls_ml2.fit(X_train, y_train)
    acc_score.append(1 - cls_ml2.score(X_test, y_test))

plt.plot(k_cnt, acc_score, 'o--', color='#eb4634')
plt.xlabel('k')
plt.ylabel('Misclassification Rate')
plt.ylim(0.10, 0.25)
plt.show()
```

# 과적합이 되지 않은 K를 찾기

```python
cls = KNeighborsClassifier(n_neighbors=6,weights='uniform')
cls.fit(X_train,y_train)

print(cls.score(X_train, y_train), 'train의 정확도')
print(cls.score(X_test, y_test),'test의 정확도')
```

# 시각화에 필요한 차원 축소 방법

```python
pca = PCA(n_components=2)
df_pca = pca.fit_transform(X_test)
df_pca = pd.DataFrame(data=df_pca,columns=['C1','C2'])
df_pca
```

```python
df_pre_x = df_pca.to_numpy()
df_pre_y = y_test['Class'].to_numpy()
```

```python
cmap_bold = ListedColormap(['#34eb92','#bb76d6'])
fig, axes = plt.subplots(1,3, figsize=(15,5))

for n_neighbors, ax in zip([1,5,20], axes):
    k_num_ml = KNeighborsClassifier(n_neighbors = n_neighbors).fit(df_pre_x,df_pre_y)
    mglearn.plots.plot_2d_separator(k_num_ml,df_pre_x, fill=True, eps=0.5, ax=ax, alpha=.3)
    mglearn.discrete_scatter(df_pre_x[:,0],df_pre_x[:,1], df_pre_y, markeredgewidth=0.1, c=['#f55182','#061a0f'], s =3, ax=ax)
    ax.set_xlabel('Com 1')
    ax.set_ylabel('Com 2')
    ax.set_title('{} Neighbors'.format(n_neighbors))

axes[0].legend(loc=3)
```

# PCA

```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
X = np.random.rand(100, 3)
```

```python
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)

plt.figure(figsize=(10,5))

ax = plt.subplot(121, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c='#72f754', marker='o')
ax.set_title('Original Data')


plt.subplot(122)
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c='#5c54f7', marker='^')

plt.title('Reduced Data by PCA')

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

plt.show()
```
