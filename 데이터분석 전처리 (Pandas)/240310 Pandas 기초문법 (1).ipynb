# 행과 열로 데이터 추출하기

- 데이터를 추출 또는 접근할 수 있다.

```python
# 열(컬럼) 기준으로 접근
# [col1]
# [[col1,col2]]
tt['age']
```

```python
# 행(인덱스) 기준으로 접근
# loc
# iloc
```

- **loc** : 행 인덱스 기준으로 추출한다. 컬럼은 문자열로도 가능! 변수의 개념으로 인덱스를 추출하는 것이지만 **-1 끝값에 대한 적용은 불가하다.**
- **iloc** : 추출한 인덱스를 기준으로 한다. 인덱스 변수로도 사용가능하며, 컬럼을 포함한 모든 것을 번호로 인식하여 지정해야 한다. **-1 끝값에 대한 적용도 가능하다!**

```python
tt.loc[9]
# 9행에 해당하는 데이터의 모든 컬럼값들을 가져옴.
```

```python
tt.iloc[9]
```

```python
tt.loc[-1]
```

```python
tt.iloc[-1]
```

```python
# 가장 첫번째 행에 해당하는 데이터를 num_row에 저장
num_row = tt.shape[0]
```

```python
# 0번째 행에서 -1을 하면 가장 마지막 행을 가져온다.
last_index = num_row -1
last_index
```

```python
# 변수의 개념으로 인덱스를 추출하는 loc
tt.loc[last_index]
```

```python
tt.iloc[last_index]
```

```python
tt.iloc[890]
```

- [시작점 : 끝점 : 증가폭 ]

```python
# 전체 데이터프레임 다 가져오기
tt.loc[::]
```

#### loc vs iloc 의 가장 큰 차이점은 컬럼에 대한 접근법이다.
- iloc : 모든 컬럼을 숫자로만 접근함.
- loc : [행,열]

```python
tt.loc[0:20:3,['who']]
```

```python
tt.iloc[0:20:3,['who']]
# 숫자가 아닌 컬럼명으로 접근했기에 오류가 난다.
```

```python
# iloc를 쓸 때는 컬럼의 인덱스(숫자)로 접근해야 한다.
tt.iloc[0:20:3,[9]]
```

```python
tt.loc[0:31:5,['fare','alive']]
```

```python
tt.iloc[0:31:5,[6,13]]
```

```python
# num_4는 일련의 숫자들이 리스트의 요소로 들어간 것이므로 숫자로 접근한 것
num_4 = list(range(0,108,4))
tt.iloc[num_4]
```

# query()
- 원하는 데이터를 추출하기 위해 추가적으로 사용하는 조건
- 요소의 특정값에 접근하여 데이터를 추출하는 케이스
- embarked = S인 경우?
- alone = True인 경우?
- fare > mean()인 경우?
- **tt.query('요구 조건')**

```python
tt.query(''' embarked == 'S' ''')
```

```python
tt.query(''' who == 'woman' ''')
```

```python
tt.query(''' who != 'woman' ''')
```

- 수치형 데이터로 접근하여 query 사용하기

```python
tt.fare.mean()
```

```python
tt['fare']
```

```python
# 평균 요금 미만인 요금을 지불한 데이터들
tt.query("fare < 32.204207968574636")
```

- 논리연산자(and, or)를 사용하여 더 복잡한 조건을 추가하기

```python
tt.query("fare >= 32.204207968574636 and age < 30")
# 평균 요금 이상을 지불했으며 나이가 30살 미만인 데이터들
```

```python
tt.query("fare < 32.204207968574636 or age < 30")
# 평균 요금 미만을 지불했거나 나이가 30살 미만인 탑승객들 정보
```

```python
tt.query('''who == 'child' and fare >= 32.204207968574636''')
# 어린아이 중 평균 요금 이상을 지불한 탑승객 정보
```

#### Q2. embarked는 S가 아니고 평균 요금 이상의 요금을 지불했으며 나이가 30살 미만인 승객들의 생존자는 몇 명일까?

```python
tt.query(''' embarked != 'S' and fare >= 32.204207968574636 and age < 30 ''')['survived'].sum()
```

```python
tt[['survived','embarked','fare','age']].query(''' embarked != 'S' and fare >= 32.204207968574636 and age < 30 ''')['survived'].sum()
```
